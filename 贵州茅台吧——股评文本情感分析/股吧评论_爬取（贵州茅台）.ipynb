```python
import requests
import pandas as pd
from lxml import etree
import csv
import time
import random
```

```python
comment_maotai_raw=[]
n=6000  #爬取6000页，80条/页
daima='600519,' #定义爬取的股票,600519为贵州茅台股票代码

for a in range(1,n):
    url='http://guba.eastmoney.com/list,'+daima+'f_'+str(a)+'.html'
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.150 Safari/537.36'}
    res = requests.get(url=url,headers=headers)
    res.encoding ='utf-8'
    tree = etree.HTML(res.text)
    li1 = tree.xpath('//*[@id="articlelistnew"]/div/span[5]/text()')
    li1.remove('发帖时间')
    li2 = tree.xpath('//*[@id="articlelistnew"]/div/span[3]/a/text()')
    
    for i in range(len(li1)):
        list1 = [li1[i],li2[i]]
        comment_maotai_raw.append(list1)
    print('第',a,'页已爬取完成')
    time.sleep(random.random()*2)
```
