```python
import requests
import pandas as pd
from lxml import etree
import csv
import time
import random
```

```python
ab1=[]
n=7271  #爬取多少页，80条/页
daima='zssh000001,' #定义爬取的股票

for a in range(20,n):
    url='http://guba.eastmoney.com/list,'+daima+'f_'+str(a)+'.html'
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.150 Safari/537.36'}
    res = requests.get(url=url,headers=headers)
    res.encoding ='utf-8'
    tree = etree.HTML(res.text)
    li1 = tree.xpath('//*[@id="articlelistnew"]/div/span[5]/text()')
    li1.remove('发帖时间')
    li2 = tree.xpath('//*[@id="articlelistnew"]/div/span[3]/a/text()')
    
    for i in range(len(li1)):
        list1 = [li1[i],li2[i]]
        ab1.append(list1)
    print('第',a,'页已爬取完成')
    time.sleep(random.random()*3)
```

```python
filename=daima+'.csv'
print(filename)
with open(filename, 'w',newline='',encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(('lastupdate_time','title'))
    writer.writerows(ab1)
```
